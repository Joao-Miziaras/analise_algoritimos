```python
import networkx as nx
import matplotlib.pyplot as plt

arestas = [(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3)]

G = nx.DiGraph()

G.add_edges_from(arestas)

nx.draw(G, with_labels=True, node_color='skyblue', node_size=1500, font_size=15, font_weight='bold', arrows=True)

plt.title('Grafo Direcionado')
plt.show()
```
