```python
import matplotlib.pyplot as plt
import mysql.connector
import pandas as pd
```

```python
import mysql.connector

def conectar_bd():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="urubu100",
        database="grupo6"
    )


```

```python
# Criar uma conexão com o banco de dados
conn = conectar_bd()

# Criar um cursor dentro do escopo da conexão
cursor = conn.cursor()

# Executar a consulta SQL
cursor.execute("SELECT temperatura, porcent_umidade, horario_da_coleta FROM dados_sensor WHERE id_sensor = 1 ORDER BY horario_da_coleta DESC LIMIT 10")

# Recuperar os dados
dados = cursor.fetchall()

df_dados = pd.DataFrame(dados, columns=['temperatura', 'umidade', 'horario'])

# Exibir os dados
print(dados)

# Fechar o cursor e a conexão
cursor.close()
conn.close()
```

```python
horarios = df_dados['horario']
temperaturas = df_dados['temperatura']
umidades = df_dados['umidade']

# Plotar os valores
plt.figure(figsize=(10, 5))
plt.plot(horarios, temperaturas, label='Temperatura')
plt.xlabel('Horário')
plt.ylabel('Valor')
plt.title('Temperatura ao longo do tempo')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(horarios, umidades, label='Umidade')
plt.xlabel('Horário')
plt.ylabel('Valor')
plt.title('Umidade ao longo do tempo')
plt.legend()
plt.show()
```
